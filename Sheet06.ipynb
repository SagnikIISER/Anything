```julia
using Dictionaries
using TensorOperations
using Random
```

```julia
function random_mps_elt(L,d,D)
      
    MPS_list = Dict();
    for i in 1:L
        if i<(L/2)
            m=d^(i-1);
            n=d^(i);
        end
        if i>(L/2)
            m=d^(L-i+1);
            n=d^(L-i);
        end
        if m>D
            m=D;
        end
        if n>D
            n=D;
        end

        MPS_list[i]=randn(d,m,n);
    end
MPS_list
end

```

```julia
A=random_mps_elt(5,2,6)
```

```julia

function create_random_mps(L,d,D)
    Lind=L;
    dind=d;
    Dind=D;
    A=random_mps_elt(Lind,dind,Dind)

    for i in 1:(L-1)
        MPS=A[i][a,b,c]*A[i+1][g,c,e];
    end
    MPS
    
end
```
