```julia
using ITensors
```

```julia
N = 100
sites = siteinds("S=1",N);
```

```julia
ampo = OpSum()
  for j=1:N-1
    ampo += "Sz",j,"Sz",j+1
    ampo += 1/2,"S+",j,"S-",j+1
    ampo += 1/2,"S-",j,"S+",j+1
  end
```

```julia
H = MPO(ampo,sites);
```

```julia
psi0 = randomMPS(sites,10);
```

```julia
sweeps = Sweeps(5)
  setmaxdim!(sweeps, 10,20,100,100,200)
  setcutoff!(sweeps, 1E-10)
```

```julia
energy, psi = dmrg(H,psi0, sweeps)
```
