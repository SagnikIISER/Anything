```julia
#The spin 1 Algebra



#The X,Y,Z gates:

g=0.00;
Z = [1 0 0; 0 0 0; 0 0 -1 ];
X = [0 1 0 ; 1 0 1; 0 1 0 ]/(sqrt(2));
Y = [0 -1im 0 ; im 0 -im; 0 im 0]/(sqrt(2));
S=  [0 1-g 0 ; 0 0 1-g; 1-g 0 0 ];

#The finite X,Y,Z Rotations:


function RX(r)
    exp(-im*r/2*X)
end


function RY(r)
    exp(-im*r/2*Y)
end


function RZ(r)
    exp(-im*r/2*Z)
end

;
```

```julia
CSV.write("X.csv",  Tables.table(X), writeheader=false)
```

```julia
o=4*pi/3;
h=rand(6)*2*pi;
J=rand(6)*2*pi;

XRow=copy(kronecker(S,S,S,S,S,S));

#XRow=copy(kronecker(RX(o),RX(o),RX(o),RX(o),RX(o),RX(o)));
ZRow=copy(kronecker(RZ(h[1]),RZ(h[2]),RZ(h[3]),RZ(h[4]),RZ(h[5]),RZ(h[6])));
```
